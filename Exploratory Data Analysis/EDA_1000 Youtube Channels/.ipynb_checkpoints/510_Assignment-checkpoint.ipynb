# #Introduction:
    This Exploratory Data Analysis deals with a data set "Most Subscribed 1000 Youtube Channels". This data set discusses about the 1000 youtube channels that are most subscribed.The columns of this data set are Rank ,Youtube channel, Subscribers, Video Views,Video Count ,Category  and Started.
    Here ,
    Rank---------------This column shows the rank on the basis of no of subscribers.
    Youtube channel----This shows the name of the youtube channel.
    Subscribers--------This column shows the number of subscribers of the youtube channel.
    Video views -------This column shows the number of views .
    Category-----------This shows the category to which the youtube channel belongs to.
    Started -----------This shows the year in which the youtube channel started.
    
    

```python
#in this codeblock we are importing the libraries required to perform EDA on this data set
#pandas to access the data from csv file and clean the data and analyze the data
#numpy to use any math when trying to analyze if required
#matplotlib to plot the stats of the data set as required.
import pandas as p
import warnings

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#in this code block we are trying to read the dataset in the folder and convert it into a dataframe named df
#df.head() displays the first five rows of the dataframe.

df = p.read_csv('dataset.csv')
df.head()
```

```python
df.tail()
#we can also use df.tail() to see what are the last 5 records of the data frame.
```

```python
df.describe()
#describe method only uses the columns that are in int data type
#so as of now there are only 2 columns as int 
```

```python
#by this command we can see what are the data types of the columns in the dataframe , and can convert if there is a need 
#we can also see the null value count
df.info()
#by this command we get to see that there are some columns that are needed to be converted into int
```

```python
df.shape
#by this command we can get to see that there are total 1000 rows and 7 columns in the data frame
```

```python
#this shows if there are any null value records if subscription column has null value
df[df['Subscribers'].isnull()] 
# Remove commas from the Subscribers column
df['Subscribers'] = df['Subscribers'].str.replace(',', '')
#converting the values to nummeric type
df['Subscribers'] = p.to_numeric(df['Subscribers'])
#converting the data type to int 
df['Subscribers']=df['Subscribers'].astype('int64')


# Remove commas from the Video Views column
#converting the values to nummeric type
#converting the data type to int 

df['Video Views'] = df['Video Views'].str.replace(',', '')
df['Video Views'] = p.to_numeric(df['Video Views'])
df['Video Views']=df['Video Views'].astype('int64')

# Remove commas from the Video Count column
#converting the values to nummeric type
#converting the data type to int 
df['Video Count'] = df['Video Count'].str.replace(',', '')
df['Video Count'] = p.to_numeric(df['Video Count'])
df['Video Count']=df['Video Count'].astype('int64')


df.info()
```

```python
#now we have converted the numeric columns that contain numbers into int
df.describe()
```

# Hypothesis Points:
1.There is coorelation between all the variables.

# Hypothesis-1-Strongest and Weakest coorelation between variables.

```python
#this method return the correlation matrix 
corr=df.corr()
#corr
```

```python
#using the seaborn package we have created a heat map to check the relations between variables or columns
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

Here we can observe that there is a coorelation between the variables.(negative or positive)
Here we are trying to identify the Strongest correlation and weakest correlation. 
The weakest correlation is observed between rank and subsribers which is -0.65, which means that if the number of subscribers increases the rank of the youtube channel decreases, which means the lowest ranked youtube channel , that is 1st has the more number of subscribers.
The strongest corelation observed is betweeen the video views and subscribers , which means subscribers and video views are directly propotional.

# Hypothesis -2-YouTube channels with a higher number of video views tend to have a higher number of subscribers

```python
#creating the plot ffigure size
plt.figure(figsize=(12,6))
#creating a scatter plot with x axis of video iews and y axis to be subcribers count in 100 million
plt.scatter(x= df['Video Views'] , y = df['Subscribers'] )
#setting the lables for x axis , y axis and also setting the title
plt.xlabel("Video Views")
plt.ylabel("Subscribers")
plt.title("Views V/S Subscribers")
```

Here we assumed that number of views could determine the number of subscriber but there are some cases where the video views are 0 but the number of subscribers are high. So in this data set there is no fixed relation between the video views and subscribers.

# Hypothesis -3-Which category has more number of channels?

```python
#this package is to ignore the warnings, so as the report can look clean
warnings.simplefilter("ignore")
#slicing the data frame as needed
df_temp=df[['Category','Youtube Channel']]
#as we are mainly analyzing the the category and channel , we are separating the channels by category name
category = df_temp.groupby('Category', as_index=False)['Youtube Channel'].count()
#category

#there are some categories which are named by a big link , so we replaced it with "others "
category['Category']= category['Category'].str.replace('https://us.youtubers.me/global/all/top-1000-most_subscribed-youtube-channels' , "Other")
category

#plotting , subplotting is done here
fig, ax = plt.subplots()
#plotting the category and number of channels in category
ax.plot(category['Category'], category['Youtube Channel'],marker='o')
#x labels , y labels and title
plt.xlabel("Name of the Category")
plt.ylabel("Number of channels")
plt.title("No of channels in a Category")

#we have rotated the lables or ticks so make the visibility clear 
xticks=plt.xticks(rotation=90)

#to avoid the text produced while using using ticks
for label in xticks[1]:
    label.set_text("")



```

Here we tried to analyze the number of channels opened in a category , and we can observe that more number of channels in youtube are from Entertainement ,Music and People's and Blogs.

# Hypothesis -4-There is growth in number of channels from 1970 to 2021

```python
#Newdf is a new dataframe which is obtained by slicing the yotube channel column and started column of the dataframe df
newdf= df[['Youtube Channel', 'Started']]
#new

#here we are re assigned the dataframe newdf to count the number of categories on the basis of year(which is present in the "started " cloumn)
newdf = newdf.groupby("Started").count()
#in the newdf there will be only one column that is yotube channel and the indexes will be the years on which it being grouped
plt.figure(figsize=(12,9))
sns.set_theme(style="whitegrid")
#as we want to create a graph between years and number of youtube channels started in that year , we are accessing the index value
#of the newdf and the column youtube channel
ax = sns.barplot(x = newdf.index , y = 'Youtube Channel' , data= newdf)
```

According to the hypothesis , we assumed that there would be a growth in the number of channels from 1970 to 2021 .But we can now observe ther was only growth till 2014, that too not at constant rate.
But after 2014 the number of youtube channels created has gradually come down from 120 to below 20.

# Hypothesis -5 -Which category has more subscribers irrespective of the channel

```python
warnings.simplefilter("ignore")
#slicing as per need into a new data frame
df_temp1=df[['Category','Subscribers']]
#df_temp1

df_temp1['Category']= df_temp1['Category'].str.replace('https://us.youtubers.me/global/all/top-1000-most_subscribed-youtube-channels' , "Other")
#df_temp1

#calculating the sum of subscribers each category has got irrespective of the channel 
df_temp1 = df_temp1.groupby("Category").sum()
df_temp1

#df_temp1.info()

fig, ax = plt.subplots()
ax.plot(df_temp1.index, df_temp1['Subscribers'],marker='o')
plt.xlabel("Name of the Category")
plt.ylabel("Number of subscribers in billions")
plt.title("No of subscribers as per category")


xticks=plt.xticks(rotation=90)

for label in xticks[1]:
    label.set_text("")



```

Here we tried to check the relation between the number of subscribers and category irrespective of the channel and we observe that music on total channels has more than 5 billion subscribers.Next to music are entertainement and peoples and blogs.

# Hypothesis-6-Which category has more number of video count

```python
warnings.simplefilter("ignore")
#slicing as per need into a new data frame
df_temp2=df[['Category','Video Count']]
df_temp2


df_temp2['Category']= df_temp2['Category'].str.replace('https://us.youtubers.me/global/all/top-1000-most_subscribed-youtube-channels' , "Other")
#df_temp1

df_temp2 = df_temp2.groupby("Category").sum()
df_temp2
#df_temp2.info()

fig, ax = plt.subplots(figsize=(10, 8))

ax.barh(df_temp2.index, df_temp2['Video Count'], align='center')

ax.set_xlabel('Video Count in Millions')
ax.set_ylabel('Category')
ax.set_title('YouTube Video Count by Category')




```

Here we have analyzed the category vs no of videos in that category , we got to see that there are more "News and Politics" category videos in youtube. I guess , irrespective of the usage of videos , news and politics videos are needed to be updated each and every day , so i think that can be the main reason , and next comes the entertainement videos.

# Hypothesis 7- Category and video views.

```python
# warnings.simplefilter("ignore")

#slicing as per need into a new data frame
df_temp3=df[['Category','Video Views']]
df_temp3

df_temp3['Category']= df_temp3['Category'].str.replace('https://us.youtubers.me/global/all/top-1000-most_subscribed-youtube-channels' , "Other")
df_temp3


df_temp3 = df_temp3.groupby("Category").sum()
df_temp3


fig, ax = plt.subplots(figsize=(10, 8))

ax.barh(df_temp3.index, df_temp3['Video Views'], align='center')

ax.set_xlabel('Video Views in 10 power 12')
ax.set_ylabel('Category')
ax.set_title('YouTube Video Count by Category')



```

By analyzing the data of category vs the total views of category irrespective of the channel , we get to see that music has most videos watched , next comes entertainment and people & blogs.

Overall, we can conclude that music and entertainment categories are the most popular on YouTube in terms of subscribers and views, while the News and Politics category has the most number of videos.
